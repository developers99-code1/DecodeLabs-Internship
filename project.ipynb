```python
import pandas as pd
#name the varaiable 
file_location = r"C:\Users\hp\Documents\Data.xlsx"
# Rea the data
df = pd.read_excel(file_location, sheet_name="Sheet1")
print("--- Data Successfully Loaded! ---")
print(df.head())
```

```python
#Phase1
# cleaning the Coupon Code
df["CouponCode"]=df["CouponCode"].fillna("No_Coupon")

#Phase :02 Duplicate Checking
#Assign the variables
dup_row=df.duplicated().sum()
dup_orders=df.duplicated(subset=['OrderID']).sum()
print(f"Audit Complete >>> Duplicated Row:{dup_row},Duplicated OrderIDS:{dup_orders}" )
#Phase 3 TEXT standarization
txt_column=["Product","PaymentMethod","OrderStatus","ReferralSource","ShippingAddress"]
for col in txt_column:
    df[col]=df[col].astype(str).str.strip().str.title()
#Date standarization
df["Date"]=pd.to_datetime(df['Date']).dt.strftime('%Y-%m-%d')
#Numeric text
df["UnitPrice"]=df["UnitPrice"].round(2)
df["TotalPrice"]=df["TotalPrice"].round(2)
print("-------Standarization done Successfully-------")
#Output :SAVING GOLD STANDARD DATA
#changing the original file with the cleaned data file
output_loc=file_location.replace("Data.xlsx","Cleaned_Data.xlsx")
df.to_excel(output_loc,index=False)
print("-------Data is now Cleaned!!--------")
```
