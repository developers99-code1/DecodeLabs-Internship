```python

import pandas as pd
import sqlite3
file_location = r"C:\Users\hp\Downloads\Data2.xlsx" #loading the data
df = pd.read_excel(file_location, sheet_name="Sheet1")#building the data base
conn = sqlite3.connect("decodelabs_db.db")
df.to_sql("sales_table", conn, if_exists="replace", index=False)

def run_sql(query):
    return pd.read_sql_query(query, conn)
print(" SQL Database Ready! Running Queries...\n")
# QUERY 1: Basic Aggregations 
print(" 1. Total Orders & Revenue Aggregation ")
q1 = """
SELECT 
    COUNT(OrderID) AS Total_Orders,
    SUM(TotalPrice) AS Grand_Total_Revenue,
    AVG(TotalPrice) AS Average_Order_Value
FROM sales_table;
"""
print(run_sql(q1))
#  QUERY 2: Product Performance 
print("\n 2. Product Performance Portfolio (GROUP BY)")
q2 = """
SELECT 
    Product,
    SUM(Quantity) AS Total_Quantity_Sold,
    SUM(TotalPrice) AS Total_Revenue
FROM sales_table
GROUP BY Product
ORDER BY Total_Revenue DESC;
"""
print(run_sql(q2))
#  QUERY 3: WHERE Clause Filtering
print("\n3. Large Shipped Orders (WHERE Clause)")
q3 = """
SELECT 
    OrderID, CustomerID, Product, Quantity, TotalPrice, OrderStatus
FROM sales_table
WHERE OrderStatus = 'Shipped' AND Quantity >= 4
LIMIT 5;
"""
print(run_sql(q3))

# QUERY 4: Payment Method Analytics 
print("\n 4. Payment Method Analytics")
q4 = """
SELECT 
    PaymentMethod,
    COUNT(OrderID) AS Transaction_Count,
    SUM(TotalPrice) AS Revenue_Contribution
FROM sales_table
GROUP BY PaymentMethod
ORDER BY Transaction_Count DESC;
"""
print(run_sql(q4))
```
